**Agglomeretive clustering and Dendogram**

Agglomeretive clustring is a bottom-up hiearchical clustring, it is same as hierchical clustring that starts with clustering approch with each data point as a single cluster

itereativly we try to create n number of clusters in every single step with respect to other data points

Dendogram is a pictorial represnetation of of how any hierarchical clustring is being conducted



The criteria or parameters used

1.Linkage criteria- how you want to merge the data points

  a.single link - min

  b.complete link - max

  c.average link - avg

  d.centroid link

  e.Wards link -we create clusters by taking minimum variance




2.Distance Metrics - the similarity or distance between data point
  
  a.Euclidean distance

  b.Manhattan Distance



3.Dendogram - Dendogram is a pictorial represnetation of of how any hierarchical clustring is being conducted



4.Stopping criteria

 a.fixed number of clusters



5.preprocessing


  

  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df=pd.read_csv("/content/cars (1) (1) (1) (1).csv")
```

```python
df
```

```python
df_copy=df.drop(["model"],axis=1)
df_copy
```

```python
# perform hierarchical clustering

Z=linkage(df_copy,method="ward")
```

```python
# create a dendrogram

plt.figure(figsize=(15,7))
dendrogram(Z,labels=np.arange(1,len(df_copy)+1))#1,33->32
plt.title("Hierarchical clustering Dendrogram")
plt.xlabel("sample index")
plt.ylabel("Distance")
plt.show()
```

```python
# model buliding

from sklearn.cluster import AgglomerativeClustering

hierarchical_cluster=AgglomerativeClustering(n_clusters=2,metric="euclidean")# affinity - which type of distance metric you want to use eg eucledian manhantan

hierarchical_cluster
```

```python
hierarchical_cluster.fit(df_copy)
```

```python
pred=hierarchical_cluster.labels_


df["class"]=pred
```
