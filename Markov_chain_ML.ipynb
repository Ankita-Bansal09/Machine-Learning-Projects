Markovify is a simple, extensible Markov chain generator. Right now, its main use is for building Markov models of large corpora of text, and generating random sentences from that. But, in theory, it could be used for other applications.

About the Dataset: This includes the entire corpus of articles published by the ABC website in the given time range. With a volume of 200 articles per day and a good focus on international news, we can be fairly certain that every event of significance has been captured here.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import random
import markovify as mv
```

```python
text_model1=mv.NewlineText(df["headline_text"],state_size=2)
```

```python
text_model2=mv.NewlineText(df["headline_text"],state_size=3)
```

```python
text_model2.make_sentence()
```

```python
text_model3=mv.NewlineText(df["headline_text"],state_size=4)
```

```python
text_model4=mv.NewlineText(df["headline_text"],state_size=3)
```

```python
combined_model=mv.combine([text_model2,text_model4],[1,1.5])
```

```python
combined_model.make_sentence()
```
